# MatPlotLib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#for command line: plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# Lesson Two

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for curent_ax in axes:
    curent_ax.plot(x,y)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('first plot')
axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

# Figure Size, Aspect Ratio & DPI

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_pic.jpg',dpi=200)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='x squared')
axes.plot(x,x**3, label='x cubed')

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

axes.legend(loc=0)
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=.4,linestyle='--',marker='*',markersize='20',
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue') #lw = linewidth, ls = linestyle
```

# Plot Range

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,lw=2,ls='--')

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
